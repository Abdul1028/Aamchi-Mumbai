```python
flight = pd.read_csv("sample_data/flight_delays.csv")
```

```python
flight.info()
```

```python

flight.set_index("Month",inplace=True)
```

```python
plt.figure(figsize=(16,6))
sns.barplot(x=flight.index,y=flight["NK"])
plt.xticks(rotation=45)
```

```python
plt.figure(figsize=(16,6))
sns.heatmap(data=flight,annot=True,cmap="Blues")
```

```python
flight_cor = flight.corr()
plt.figure(figsize=(16,6))
sns.heatmap(data=flight_cor,annot=True,cmap="pink")
```

```python
import pandas as pd
insurance =pd.read_csv("sample_data/insurance.csv")
insurance
```

```python
import seaborn as sns
sns.scatterplot(x=insurance['bmi'],y=insurance['charges'])
```

```python
sns.scatterplot(x=insurance['bmi'],y=insurance['charges'],hue=insurance['smoker'])
```

```python
sns.regplot(x=insurance['bmi'],y=insurance['charges'])
```

```python
sns.lmplot(x="bmi",y="charges",hue="smoker",data=insurance)
```

```python
sns.swarmplot(x=insurance['smoker'],y=insurance['charges'])
```

```python
sns.stripplot(x=insurance['smoker'],y=insurance['charges'],jitter=True)
```
